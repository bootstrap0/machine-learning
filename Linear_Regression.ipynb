```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

df = pd.read_csv('data.csv')

# Write your code below
# defining the explanatory (independent) and response (dependent) variables
# defining the independent variable
ind_vars1 = ['annual_income_in_usd']
X = df.drop('annual_expenditure', axis=1)
y = df['annual_expenditure']

# splitting the data in 80:20 ratio for train and test sets
X_train, X_test, y_train, y_test = train_test_split(
    X,    # specifying the independent variables
    y,    # specifying the dependent variable
    test_size=0.20,    # specifying the size of the test set as a fraction of the whole data
    random_state=42    # specifying a seed value to enable reproducible results
)

# creating an instance of the linear regression model
model = LinearRegression()

# fitting the model to the training data
model.fit(X_train[ind_vars1], y_train)

y_pred = model.predict(X_test[ind_vars1])
r_squared = r2_score(y_test, y_pred)  # to compute R-squared
```

```python
import pandas as pd
import numpy as np

# to visualize data
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression

# to split the data into train and test sets
from sklearn.model_selection import train_test_split
# to check a regression model's performance
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

df = pd.read_csv('https://raw.githubusercontent.com/bootstrap0/machine-learning/refs/heads/main/insurance.csv')
print(df.shape)
print(df.isnull().sum())
sns.boxplot(data=df, y='charges', x='region');
df['bmi_log'] = np.log(df.bmi)
sns.histplot(data=df, x='bmi_log', kde=True)

df.groupby('region').charges.median()

df.corr(numeric_only=True)

df = pd.get_dummies(df, columns=['sex','smoker','region'], drop_first=True)
df.head()

X = df.drop(['charges', 'bmi_log'],axis=1)
y = df['charges']
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)
x_test.shape

lr = LinearRegression()
lr.fit(x_train, y_train)
print("Intercept:", lr.intercept_)

pred_train = lr.predict(x_train)
r2 = r2_score(y_train, pred_train)
pred_test = lr.predict(x_test)
rmse = np.sqrt(mean_squared_error(y_test, pred_test))
mae = mean_absolute_error(y_test, pred_test)
print(mae)
print(r2)
print(rmse)
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split

df = pd.read_csv('data.csv')

# Define and train the model
X = df[['Product Price']]
y = df['Sales']


# splitting the data in 80:20 ratio for train and test sets
X_train, X_test, y_train, y_test = train_test_split(
    X,    # specifying the independent variables
    y,    # specifying the dependent variable
    test_size=0.30,    # specifying the size of the test set as a fraction of the whole data
    random_state=1    # specifying a seed value to enable reproducible results
)

lr = LinearRegression()
lr.fit(X_train, y_train)

y_pred_train = lr.predict(X_train)
y_pred_test = lr.predict(X_test)

# Predict and calculate Adjusted R²
mse_train = mean_squared_error(y_train, y_pred_train)
mse_test = mean_squared_error(y_test, y_pred_test)
```
